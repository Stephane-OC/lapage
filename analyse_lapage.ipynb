# üéØ Objectif du projet ‚Äî Lapage üìö

Ce notebook a pour objectif d‚Äô**analyser les ventes et le comportement des clients de l‚Äôentreprise Lapage**, une librairie ayant d√©velopp√© son activit√© de vente en ligne, afin d‚Äôen extraire des **insights clairs et exploitables pour le comit√© de direction (CODIR)**.

L‚Äôanalyse vise √† fournir une **vision synth√©tique et strat√©gique de la performance commerciale**, ainsi qu‚Äôune meilleure compr√©hension des profils et habitudes des clients.

---

## üß© Contexte

Initialement positionn√©e comme une **librairie physique**, l‚Äôentreprise **Lapage** a lanc√© son site de vente en ligne il y a deux ans afin de r√©pondre √† la demande croissante de ses clients et d‚Äô√©largir son activit√©.

Dans ce contexte, la direction souhaite aujourd‚Äôhui :

- mieux **suivre l‚Äô√©volution de l‚Äôactivit√© e-commerce** ;
- comprendre les **dynamiques de vente** ;
- analyser le **comportement des clients en ligne**.

Pour mener cette √©tude, trois jeux de donn√©es sont mis √† disposition :

- un fichier **clients** (genre, ann√©e de naissance) ;
- un fichier **produits** (prix et cat√©gories) ;
- un fichier **transactions** (dates d‚Äôachat, clients et produits).

---

## üìà Objectifs principaux

Le travail se structure autour de **deux axes compl√©mentaires**.

### Axe 1 ‚Äî Analyse de l‚Äôactivit√© commerciale

- √âtudier l‚Äô**√©volution du chiffre d‚Äôaffaires dans le temps** ;
- Mettre en √©vidence les tendances √† l‚Äôaide de **moyennes mobiles** ;
- Analyser le **nombre de clients**, de **transactions** et de **produits vendus** ;
- Identifier les **produits les plus performants (tops)** et les **moins performants (flops)** ;
- √âtudier la **r√©partition des ventes par cat√©gorie**.

---

### Axe 2 ‚Äî Analyse du comportement client

- √âtudier la **r√©partition du chiffre d‚Äôaffaires par client** ;
- Analyser la **dispersion des ventes** √† l‚Äôaide d‚Äôune **courbe de Lorenz** ;
- Mettre en √©vidence des **corr√©lations entre les variables clients** :
  - √¢ge ;
  - genre ;
  - fr√©quence d‚Äôachat ;
  - panier moyen ;
  - cat√©gories de livres achet√©es.
- Appliquer et justifier l‚Äôutilisation de **tests statistiques adapt√©s** :
  - test du **œá¬≤ (khi-deux)** ;
  - **ANOVA** ou **test de Student** ;
  - **corr√©lation de Pearson ou de Spearman**.
- Identifier et traiter les **clients atypiques (BtoB)** afin de garantir la pertinence des analyses.

---

## üß† Finalit√©

Les r√©sultats issus de cette √©tude permettront :

- d‚Äô**accompagner la prise de d√©cision strat√©gique du CODIR** ;
- d‚Äôorienter les **choix marketing et commerciaux** (offres, ciblage, positionnement) ;
